# SORTING

```
import pandas as pd
import numpy as np
```

```
#read data set
data=pd.read_csv("/content/bigmart_data.csv", sep=',')
#drop the null values
data=data.dropna(how='any')
data.head()
```

## Sorting

sort_values[]- by columns  and sort_index[]- by rows


```
sortdata1=data.sort_values(by='Outlet_Establishment_Year')
```

```
sortdata1 #gives whole excel to get specific rows, say first 5 rows give sortdata1[:5]
```

```
sortdata1[:5]
```

```
data2=data
```

```
#now if we want to sort the data in descending format and want the change to be made in the original database
#and not store in new variable--- use inplace

data2.head()
```

```
data2.sort_values(by='Outlet_Establishment_Year', ascending=False, inplace=True)
```

```
data2.head()
```

```
#now if we want to so sort based on multiple columns then we have to provide in form of list
#here sort based on first variable and then sort on the basis of second variable

data.sort_values(by=['Outlet_Establishment_Year','Item_Outlet_Sales'], ascending=False)[:5]
```

```
data.sort_values(by=['Item_Outlet_Sales', 'Outlet_Establishment_Year'], ascending=False, inplace=True)
```

```
data[:5]
```

```
#now every row index is mixed up. If want to retrieve original data

data.sort_index(inplace=True)
```

# MERGING DATAFRAMES

-MERGE
-CONCAT

```
#CREATING DUMMY DATA
df1=pd.DataFrame({'A':['A0','A1','A2','A3'],
                  'B':['B0','B1','B2','B3'],
                  'C':['C0','C1','C2','C3'],
                  'D':['D0','D1','D2','D3']},
                 index=[0,1,2,3])
```

```
df2=pd.DataFrame({'A':['A4','A5','A6','A7'],
                  'B':['B4','B5','B6','B7'],
                  'C':['C4','C5','C6','C7'],
                  'D':['D4','D5','D6','D7']},
                 index=[4,5,6,7])
```

```
df3=pd.DataFrame({'A':['A8','A9','A10','A11'],
                  'B':['B8','B9','B10','B11'],
                  'C':['C8','C9','C10','C11'],
                  'D':['D8','D9','D10','D11']},
                 index=[8,9,10,11])
```

```
df1
```

```
df2
```

```
df3
```

```
#concat rowwise
result=pd.concat([df1,df2,df3])
```

```
result
```

```
#now if we want to give index to each dataframe

result=pd.concat([df1,df2,df3], keys=['df1','df2','df3']) #by default this is outer join
result
```

```
result.loc['df2']
```

```
#concat column wise - use axis in the concat function

df4=pd.DataFrame({'B':['B2','B3','B6','B7'],
                  'D':['D2','D3','D6','D7'],
                  'E':['E2','E3','E6','E7']},index=[2,3,6,7])
```

```
df4
```

```
result2=pd.concat([df1,df4],axis=1,sort=False) #default- outer join
result2
```

```
result2=pd.concat([df1,df4],sort=False) # axis not use- thus row wise merged
result2
```

```
result2=pd.concat([df1,df4],axis=1,join='inner') # join=inner gives just the intersection
result2
```
